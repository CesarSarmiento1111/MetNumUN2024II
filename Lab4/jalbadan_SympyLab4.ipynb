## SymPy’s Symbolic and Numercical Complex Evaluations
- https://docs.sympy.org/latest/modules/evalf.html](https://)
- x = x1 + I*x2,y = y1 + I*y2, z = z1 + I*z2, x * y * z

a= i*a1**2 + (i**3*a2**3+2), b = i*b1 + 2*b2, c = i*c1 + c2, a*b*c

image.png

```
from sympy import symbols, I, N, pi, expand
#a= i*a1**2 + (i**3*a2**3+2), b = i*b1 + 2*b2, c = i*c1 + c2, a*b*c
a1, a2, b1, b2, c1, c2 = symbols("a1 a2 b1 b2 c1 c2", real=True)
a = I*a1**2 + I**3*a2**3+2
b = I*b1 + 2*b2
c = I*c1 + c2

print(a*b*c)
print(expand(a*b*c))
print(expand((a*b)*c))
print(expand(a*(b*c)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
